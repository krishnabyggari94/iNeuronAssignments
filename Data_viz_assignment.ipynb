```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris,target=datasets.load_iris(return_X_y=True);
iris.shape,target.shape
```

```python
pca=decomposition.PCA(n_components=3,random_state=42)
trans_iris=pca.fit_transform(iris)
print(trans_iris.shape)
```

```python
fig=plt.figure(1,figsize=(6,6))
axis=Axes3D(fig)
axis.scatter(xs=trans_iris.T[0],ys=trans_iris.T[1],zs=trans_iris.T[2],c=target)
```
